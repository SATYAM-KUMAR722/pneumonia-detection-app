```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "chest-xray-pneumonia/chest_xray/train",
    labels="inferred",
    label_mode="int",
    image_size=(256, 256),
    color_mode="rgb",
    batch_size=32,
    shuffle=True
)
```

```python
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "chest-xray-pneumonia/chest_xray/test",
    labels="inferred",
    label_mode="int",
    image_size=(256, 256),
    color_mode="rgb",
    batch_size=32,
    shuffle=True
)
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "chest-xray-pneumonia/chest_xray/val",
    labels="inferred",
    label_mode="int",
    image_size=(256, 256),
    color_mode="rgb",
    batch_size=32,
    shuffle=True
)
```

```python
from tensorflow.keras.layers import (
    Input, Rescaling, RandomFlip, RandomRotation,
    RandomZoom, RandomContrast)
```

```python
data_augmentation = tf.keras.Sequential([
    RandomFlip("horizontal"),
    RandomRotation(0.1),
    RandomZoom(0.1),
    RandomContrast(0.15)
])
```

```python
# FEATURE EXTRACTION
from tensorflow.keras.layers import (
    Input, Rescaling, RandomFlip, RandomRotation,
    Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
)
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Sequential

base_model = DenseNet121(
    weights="imagenet",
    include_top=False,
    input_shape=(256, 256, 3)
)

base_model.trainable = False


```

```python
model = Sequential([

    Input(shape=(256, 256, 3)),

    Rescaling(1./255),
    RandomRotation(0.1),

    base_model,

    GlobalAveragePooling2D(),
    BatchNormalization(),

    Dense(128, activation="relu"),
    Dropout(0.3),

    Dense(1, activation="sigmoid")
])
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
callbacks = [
    tf.keras.callbacks.ModelCheckpoint(
        "best_densenet_feature_extraction.keras",
        monitor="val_loss",
        save_best_only=True,
        verbose=1
    ),

    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=5,
        restore_best_weights=True,
        verbose=1
    ),

    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.3,
        patience=2,
        min_lr=1e-6,
        verbose=1
    )
]
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,
    callbacks=callbacks
)
```

```python
model.load_weights("best_densenet_feature_extraction.keras")
```

```python
# Fine_tuning

base_model.trainable = True

for layer in base_model.layers[:-40]:
    layer.trainable = False
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
history_finetune = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=callbacks
)

```

```python
model.load_weights("best_densenet_feature_extraction.keras")
```

```python
y_true, y_pred = [], []

for images, labels in test_ds:
    preds = model.predict(images)
    preds = (preds > 0.5).astype("int32")

    y_pred.extend(preds.flatten())
    y_true.extend(labels.numpy())
```

```python
from sklearn.metrics import accuracy_score
print("Test Accuracy:", accuracy_score(y_true, y_pred) * 100)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_true, y_pred)

sns.heatmap(
    cm, annot=True, fmt="d", cmap="Blues",
    xticklabels=["Normal", "Pneumonia"],
    yticklabels=["Normal", "Pneumonia"]
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
